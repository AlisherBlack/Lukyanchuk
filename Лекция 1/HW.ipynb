```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 10.6e-6
n = 12.8
k = 52.1

def denominator(h):
    pi = np.pi
    return pi**2 * (n**2 + k**2)**2 * (h/l)**2 + 4*pi*n*k*h/l + 1
    
def R(h):
    pi = np.pi
    return (pi**2 * (n**2 + k**2)**2 * (h/l)**2) / denominator(h)

def T(h):
    pi = np.pi
    numerator = 4 * pi**2 * (n**2 + k**2)*(h/l)**2 + 4*pi*k*h/l + 1
    return numerator / denominator(h)

def A(h):
    return 1-R(h)-T(h)

def new_A(h):
    pi = np.pi
    numerator = -4 * pi**2 * (n**2 + k**2)*(h/l)**2 + 4*pi*k*h/l - 4*pi*k*h/l
    return numerator / denominator(h)
    
```

```python
x = np.linspace(0, 50, 10000)*1e-9

plt.plot(x, A(x))
```

```python
x = np.linspace(0, 50, 10000)*1e-9

plt.plot(x, new_A(x))
```
